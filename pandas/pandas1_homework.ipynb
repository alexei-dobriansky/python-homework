### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('pandas_data/power.csv')
```

Суммарно по всем категориям:

```python
def baltic(country):
    if country in ['Latvia', 'Lithuania', 'Estonia']:
        return True
    return False

def suitable_category(category):
    if category in [4, 12, 21]:
        return True
    return False

def accounting_period(year):
    return 2005 <= year <= 2010

data['is_Baltic'] = data.country.apply(baltic)
data['is_suitable_cat'] = data.category.apply(suitable_category)

baltic_data = data[ data['is_Baltic'] & data['is_suitable_cat'] & (data.quantity > 0) & (data.year.apply(accounting_period)) ]
baltic_data['quantity'].sum()
```

По каждой категории:

```python
baltic_data.groupby('category').sum()['quantity'].to_frame()
```
