### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
movies = pd.read_csv('pandas_data_2/movies.csv', sep=',')
ratings = pd.read_csv('pandas_data_2/ratings.csv', sep=',')
```

```python
movies_rating = ratings.groupby('movieId').mean().reset_index()[['movieId','rating']]
movies_with_rating = movies.merge(movies_rating, on='movieId', how='left')

def classification(rating):
    if rating <= 2:
        return 'Низкий'
    elif rating <= 4:
        return 'Средний'
    elif rating > 4:
        return 'Высокий'
    else:
        return 'Без рейтинга'

movies_with_rating['class'] = movies_with_rating.rating.apply(classification)
movies_with_rating.head()
```

## Задание 2
Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
keywords = pd.read_csv('pandas_data_2/keywords.csv', sep=',')
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def find_region(keyword):
    for region, city_list in geo_data.items():
        for city in city_list:
            if city in keyword:
                return region
    return 'undefined'            

keywords['region'] = keywords.keyword.apply(find_region)
```

```python
keywords.query('region in ("Центр", "Северо-Запад", "Дальний Восток")')
```

## Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия, т.е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

```python
years = list(range(1950, 2011))
```

```python
movies = pd.read_csv('pandas_data_2/movies.csv', sep=',')
ratings = pd.read_csv('pandas_data_2/ratings.csv', sep=',')
```

```python
def production_year(title):
    for year in years:
        if str(year) in title:
            return year
    return 1900

def get_rating(movieId):
    return ratings[ratings['movieId'] == movieId].rating.mean()

movies['year'] = movies.title.apply(production_year)
movies['rating'] = movies.movieId.apply(get_rating)

movies.groupby('year').mean().sort_values('rating',ascending=False)
```

Вывод: раньше снимали настоящее кино, не то что сейчас!
