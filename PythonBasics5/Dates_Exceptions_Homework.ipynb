### Задание 1.
Печатные газеты использовали свой формат дат для каждого выпуска.  
Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:  
The Moscow Times - Wednesday, October 2, 2002  
The Guardian - Friday, 11.10.13  
Daily News - Thursday, 18 August 1977  

```python
from datetime import datetime

dict_format = {}
dict_format["The Moscow Times"] = "%A, %B %d, %Y"
dict_format["The Guardian"] = "%A, %d.%m.%y"
dict_format["Daily News"] = "%A, %d %B %Y"

print("The Moscow Times:",datetime.strptime('Wednesday, October 2, 2002', dict_format["The Moscow Times"]))
print("The Guardian:",datetime.strptime('Friday, 11.10.13', dict_format["The Guardian"]))
print("Daily News:",datetime.strptime('Thursday, 18 August 1977', dict_format["Daily News"]))
```

### Задание 2
Дан поток дат в формате YYYY-MM-DD, в которых встречаются некорректные значения:  
stream = [‘2018-04-02’, ‘2018-02-29’, ‘2018-19-02’]  
Напишите функцию, которая проверяет эти даты на корректность.  
Т. е. для каждой даты возвращает True (дата корректна) или False (некорректная дата).

```python
from datetime import datetime

stream = ['2018-04-02', '2018-02-29', '2018-19-02']

def check_date(date):
    try:
        datetime.strptime(date, '%Y-%m-%d')
        return True
    except:
        return False

for date in stream:
    date_status = lambda date: 'корректная' if check_date(date) else 'некорректная'
    print(f'Дата {date} {date_status(date)}')
```

### Задание 3
Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date.  
Даты должны вводиться в формате YYYY-MM-DD. В случае неверного формата или при start_date > end_date должен возвращаться пустой список.

```python
from datetime import datetime, timedelta

start_date = '2021-02-01'
end_date = '2021-02-05'

def date_range(start_date, end_date):
    date_format = '%Y-%m-%d'
    list_of_dates = []
    try:
        start_datetime = datetime.strptime(start_date, date_format)
        end_datetime = datetime.strptime(end_date, date_format)
        
        for i in range((end_datetime - start_datetime).days + 1):
            list_of_dates.append(datetime.strftime(start_datetime + timedelta(days=i), date_format))
        
        return list_of_dates
    except:
        return []
    
for date in date_range(start_date, end_date):
    print(date)
```
