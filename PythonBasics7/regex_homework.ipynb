### Задание 1
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

```python
import re

def validate_car_id(car_id):
    region = r'[17]?\d{2}$'
    pattern = r'[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ][АВЕКМНОРСТУХ]'
    
    if re.match(pattern + region, car_id):
        print('Номер {} валиден, регион {}'.format(car_id[0:6], re.search(region, car_id).group()))
    else:
        print('Номер не валиден')
        
validate_car_id('А222УА177')
```

### Задание 2
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
def remove_repeated_words(some_string):
    return re.sub(r'(\w+)( \1)+', r'\1', some_string)

some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять \
все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'

remove_repeated_words(some_string)
```

### Задание 3
Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

```python
def acronym(some_words):
    pattern = r'^\w|(?<=\s)\w'
    return ''.join(re.findall(pattern, some_words))

acronym('Near Field Communication')
```

### Задание 4
Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

```python
def print_domains(email_list):
    domains = re.findall(r'(?<=@)\w+\.\w+', ','.join(email_list))
    for domain in set(domains):
        print(domain, domains.count(domain))
        
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
print_domains(emails)
```

### Задание 5
Напишите функцию, которая будет подсчитывать сколько слов начинается на гласные, а сколько на согласные буквы в тексте (текст может быть написан как с использованием букв кириллицы, так и латиницы).

```python
def vowels_n_consonants(some_text):
    vowels = r'(?<=\b)[аеёиоуыэюяАЕЁИОУЫЭЮЯaeiouyAEIOUY]'
    consonants = r'(?<=\b)[бвгджзйклмнпрстфхцчшщБВГДЖЗЙКЛМНПРСТФХЦЧШЩbcdfghjklmnpqrstvwxzBCDFGHJKLMNPQRSTVWXZ]'
    print('Слов на гласные буквы', len(re.findall(vowels, some_text)))
    print('Слов на согласные буквы', len(re.findall(consonants, some_text)))

some_text = 'Эталонной реализацией Python является интерпретатор CPython, \
поддерживающий большинство активно используемых платформ. \
Он распространяется под свободной лицензией Python Software Foundation License, \
позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'
vowels_n_consonants(some_text)
```

### Задание 6
Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:  
+7-xxx-xxx-xx-xx

```python
def verify_phone_no(phone_no):
    phone_only_digits = re.sub(r'\D', r'', phone_no)
    pattern = r'^[78](\d{3})(\d{3})(\d{2})(\d{2})$'

    if re.match(pattern, phone_only_digits):
        print(re.sub(pattern, r'+7-\1-\2-\3-\4', phone_only_digits))
    else:
        print('Номер не валиден')
        
verify_phone_no('+7 955 555-55-55')
verify_phone_no('8(955)555-55-55')
verify_phone_no('+7 955 555 55 55')
verify_phone_no('8(955) 555-55-55')
verify_phone_no('423-555-55-5555')
verify_phone_no('123-456-789')
```
