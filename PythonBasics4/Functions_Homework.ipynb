## Домашнее задание "Функции"

### 1. Основное задание

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def create_list_of_docs(docs, shelves):
    '''
    создание списка документов с указанием мест их хранения
    '''
    list_of_docs = []
    for doc in docs:
        current_doc = doc.copy()
        for shelf in shelves:
            if current_doc['number'] in shelves[shelf]:
                current_doc.update({'directory': shelf})
                list_of_docs.append(current_doc)
    return list_of_docs

def print_document_owner(list_of_docs):
    document_no = input('Введите номер документа: ')
    found_data = find_in_documents(list_of_docs, 'name', document_no)
    if found_data != '':
        print('Владелец документа:', found_data)
    else:
        print('Документ не найден в базе')

def print_document_location(list_of_docs):
    document_no = input('Введите номер документа: ')
    found_data = find_in_documents(list_of_docs, 'directory', document_no)
    if found_data != '':
        print('Документ хранится на полке:', found_data)
    else:
        print('Документ не найден в базе')

def print_all(list_of_docs):
    for doc in list_of_docs:
        print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {doc['directory']}")

def find_in_documents(list_of_docs, what, document_no):
    found_data = ''
    for doc in list_of_docs:
        if ('number', document_no) in doc.items():
            found_data = doc[what]
            break
    return found_data

def add_shelf():
    shelf_no = input('Введите номер полки: ')
    if shelf_no not in directories.keys():
        directories.update({shelf_no: []})
        print('Полка добавлена')
    else:
        print('Такая полка уже существует')

def delete_shelf():
    shelf_no = input('Введите номер полки: ')
    if shelf_no in directories.keys():
        if directories[shelf_no] != []:
            print('На полке есть документа, удалите их перед удалением полки')
        else:
            del(directories[shelf_no])
            print('Полка удалена.')
    else:
        print('Такой полки не существует.')
    
def print_all_locations():
    print('Текущий перечень полок:', ', '.join(list(directories.keys())))

def main():
    list_of_docs = create_list_of_docs(documents, directories)
    while True:
        user_command = input("Введите команду: ")
        if user_command == 'p':
            print_document_owner(list_of_docs)    
        elif user_command == 's':
            print_document_location(list_of_docs)
        elif user_command == 'l':
            print_all(list_of_docs)
        elif user_command == 'as':
            add_shelf()
            print_all_locations()
        elif user_command == 'ds':
            delete_shelf()
            print_all_locations()
        elif user_command == 'q':
            print('Конец программы')
            break
        else:
            print('Неизвестная команда')
    return

main()
```
