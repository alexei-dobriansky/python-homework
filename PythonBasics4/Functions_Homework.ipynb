## Домашнее задание "Функции"

### 1. Основное задание

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

# команды
def print_document_owner():
    '''
    поиск владельца документа, команда p
    '''
    document_no = input('Введите номер документа: ')
    current_doc = find_in_documents(documents, document_no)
    if current_doc != {}:
        print('Владелец документа:', current_doc['name'])
    else:
        print('Документ не найден в базе')

def print_document_location():
    '''
    поиск полки хранения документа, команда s
    '''
    document_no = input('Введите номер документа: ')
    current_doc = find_in_documents(create_list_of_docs(), document_no)
    if current_doc != {}:
        print('Документ хранится на полке:', current_doc['directory'])
    else:
        print('Документ не найден в базе')

def print_all():
    '''
    печать всей информации о документах, команда l
    '''
    print('Текущий список документов:')
    for doc in create_list_of_docs():
        print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {doc['directory']}")

def add_document():
    '''
    добавление документа в список, команда ad
    '''
    document_no = input('Введите номер документа: ')
    document_type = input('Введите тип документа: ')
    document_owner = input('Введите владельца документа: ')
    shelf_no = input('Введите полку для хранения: ')
    if check_shelf(shelf_no):
        if find_in_documents(documents,document_no) == {}:
            documents.append(dict({'type': document_type, 'number': document_no, 'name': document_owner}))
            directories[shelf_no].append(document_no)
            print('Документ добавлен')
            print_all()
        else:
            print('Такой документ уже существует')

def delete_document():
    '''
    удаление документа, команда d
    '''
    document_no = input('Введите номер документа: ')
    current_doc = find_in_documents(documents, document_no)
    current_doc_with_shelf = find_in_documents(create_list_of_docs(), document_no)
    if current_doc != {}:
        documents.remove(current_doc)
        directories[current_doc_with_shelf['directory']].remove(document_no)
        print('Документ удален')
        print_all()
    else:
        print('Документ не найден в базе')

def move_document():
    '''
    перемещение документа, команда m
    '''
    document_no = input('Введите номер документа: ')
    shelf_no = input('Введите полку для хранения: ')
    if check_shelf(shelf_no):
        current_doc = find_in_documents(create_list_of_docs(), document_no)
        if current_doc != {}:
            directories[current_doc['directory']].remove(document_no)
            directories[shelf_no].append(document_no)
            print('Документ перемещен')
            print_all()
        else:
            print('Документ не найден в базе')
 
def add_shelf():
    '''
    добавление полки, команда as
    '''
    shelf_no = input('Введите номер полки: ')
    if shelf_no not in directories.keys():
        directories.update({shelf_no: []})
        print('Полка добавлена')
    else:
        print('Такая полка уже существует')

def delete_shelf():
    '''
    удаление полки, команда ds
    '''
    shelf_no = input('Введите номер полки: ')
    if check_shelf(shelf_no):
        if directories[shelf_no] != []:
            print('На полке есть документы, удалите их перед удалением полки')
        else:
            del(directories[shelf_no])
            print('Полка удалена.')


# служебные функции
def create_list_of_docs():
    '''
    создание списка документов с указанием мест их хранения
    '''
    list_of_docs = []
    for doc in documents:
        current_doc = doc.copy()
        for shelf in directories:
            if current_doc['number'] in directories[shelf]:
                current_doc.update({'directory': shelf})
                list_of_docs.append(current_doc)
    return list_of_docs

def find_in_documents(list_of_docs, document_no):
    '''
    поиск документа в справочнике
    '''
    found_doc = {}
    for doc in list_of_docs:
        if ('number', document_no) in doc.items():
            found_doc = doc
            break
    return found_doc

def check_shelf(shelf_no):
    if shelf_no in directories.keys():
        return True
    else:
        print('Такой полки не существует. Добавьте полку командой as')
        return False

def print_all_locations():
    print('Текущий перечень полок:', ', '.join(list(directories.keys())))


# тело программы
def main():
    while True:
        user_command = input("Введите команду: ")
        if user_command == 'p':
            print_document_owner()    
        elif user_command == 's':
            print_document_location()
        elif user_command == 'l':
            print_all()
        elif user_command == 'ad':
            add_document()
        elif user_command == 'd':
            delete_document()
        elif user_command == 'm':
            move_document()
        elif user_command == 'as':
            add_shelf()
            print_all_locations()
        elif user_command == 'ds':
            delete_shelf()
            print_all_locations()
        elif user_command == 'q':
            print('Конец программы')
            break
        else:
            print('Неизвестная команда', user_command)
    return

main()
```
