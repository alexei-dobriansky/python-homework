## Домашнее задание к лекции «Случайные события. Случайные величины»

Смоделировать игру против лотерейного автомата типа “777”.   
Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999.  
При некоторых значениях числа игрок получает выигрыш (см. таблицу)  
Выгодна ли игра игроку? Сколько в среднем приобретает или теряет игрок за одну игру?

| Комбинация | Выигрыш |
|-----|----|
| 777 | 200 руб.|
| 999 | 100 руб.|
| 555 | 50 руб.|
| 333 | 15 руб.|
| 111 | 10 руб.|
| *77 | 5 руб.|
| **7 | 3 руб.|
| *00 | 2 руб.|
| **0 | 1 руб.|

```python
import numpy as np
```

```python
np.random.seed(3)

winning_comb = {'777': 200, '999': 100, '555': 50, '333': 15, '111': 10, '77': 5, '7': 3, '00': 2, '0': 1}

no_of_games = 500000
total_winnings = 0

for i in range(no_of_games):
    total_winnings -= 1

    comb_str = ''.join([str(x) for x in np.random.randint(0, 10, 3)])
        
    while comb_str:
        if winning_comb.get(comb_str):
            total_winnings += winning_comb[comb_str]
            break
        comb_str = comb_str[1:]
            
print('Средний выигрыш на одну игру:', total_winnings / no_of_games)
    
```
